## Goal
The goal of encoding is to find a mapping function $\phi$ that for every feature vector $x$ in the dataset $\mathcal{D}$ gives a representation of the feature vector in the quantum world. This mapping function can then be translated in a unitary to use as a gate in our circuit, and in the parameters of our models.


## Basis encoding

Most straight-forward approach :  

Dataset : $$\mathcal{D} = (x^1,...,x^M) \text{ , with } M\in\mathbb{N}^*$$ 
Feature vector : $$\forall i \in [1,M], x^i = (b_1,...,b_N) \text{ , with } \forall j \in [1,N], b_j \in \{0,1\}$$
Mapping function : $$\phi : x^i \rightarrow |x^i\rangle$$
Full encoding : $$\Phi(\mathcal{D}) = \frac{1}{\sqrt{M}}\sum_{i = 1}^M \phi(x^i) = \frac{1}{\sqrt{M}}\sum_{i=1}^M |x^i\rangle$$
Most adapted when the features of the feature vector are binary.  

Not memory efficient : for a dataset composed of $M$ feature vector that have $N$ features, the amount of qubit required is $NM$

### Example 
Hence we want to encode the following dataset : 
$$\mathcal{D} = (x^1,x^2) \text{ with } x^1 = (1) \text{ and } x^2 = (3)$$

The encoding of $\mathcal{D}$ would be : 
$$\Phi(\mathcal{D}) = \frac{1}{\sqrt{2}}(|01\rangle + |11\rangle)$$

In Qiskit, this type of encoding can be realized using the *initialize* method of the *QuantumCircuit* class :


```python
backend = Aer.get_backend('statevector_simulator')
```

```python
qc = QuantumCircuit(2)#We need 2 qubits to encode 1 and 4
qc.initialize([
    0,
    1/math.sqrt(2),#|01>
    0,
    1/math.sqrt(2)#|11>
])
#We can send our circuit to the state_vector simulator to check if our
#encoding is correct
execute(qc,backend).result().get_statevector().draw(output = 'latex')
```

## Amplitude encoding

Contrary to basis encoding, amplitude encoding can encode any feature vector $x$ such that : 
$$x^m = (x_1^m,...,x_N^m) \text{ with }x_i^m \in \mathbb{R}, \forall i \in [1,N], m \in [1,M] \text{ , M the number of feature vector in the dataset}$$
A condition is however required on $x$ :
$$||x|| = 1$$
That can be achieved by simply normalizing x.  

Mapping function : 
$$\phi : x_i  \rightarrow x_i|i\rangle, \forall i \in [1,N]$$
Full encoding : 
$$\Phi(\mathcal{D}) = \sum_{m = 1}^M\sum_{i = 1}^N x_i^m|i\rangle|m\rangle $$
This representation is more memory efficient than the basis encoding. Indeed, we only need $log_2(NM)$ qubits to encode the dataset.

### Example
Hence we want to encode the following dataset which contains real values :
$$\mathcal{D} = (x^1,x^2) \text{ , with } x^1 = (1.5,3) \text{ and } x^2 = (-3.6, 8.99)$$
The encoding of $\mathcal{D}$ would be : 
$$(x^1,x^2)_{normalized} = \frac{1}{\sqrt{105.0301}}(1.5,3,-3.6,8.99)$$
$$\Phi(\mathcal{D}) = \frac{1}{\sqrt{105.0301}}( 1.5|00\rangle + 3|01\rangle - 3.6|10\rangle + 8.99|11\rangle)$$
In Qiskit, this type of encoding is also realized using the *initialize* method :

```python
import IPython
x = [1.5,3,-3.6,8.99]
n = norm(x)
x /= n#We normalize the vector to give it to the initialize method
qc = QuantumCircuit(2)
qc.initialize(x)
#After execution, we multiply our statevector by the norm for the reader to recognize 
#the statevector computed above.
display(((execute(qc,backend).result().get_statevector())*n).draw(output = 'latex'))
print("To see the real statevector, this result should be divided by",n)
```

## Angle encoding

Another way of encoding a feature vector is to use the angle of a feature. Hence we have $x = (x_1,...,x_n)$ a feature vector with $\forall i \in [1,N], x_i \in \mathbb{R}$, then we can encode $x$ using the following mapping function : 
$$\phi : x_i \rightarrow cos(x_i)|0\rangle + sin(x_i)|1\rangle$$
Thus, for the feature vector, we have the mapping function : 
$$\phi' : x \rightarrow \bigotimes_{i=1}^N cos(x_i)|0\rangle + sin(x_i)|1\rangle$$

A huge advantage of angle encoding is that we can transform the function $\phi'$ into an unitary, indeed : 
$$\phi(x_i) \Leftrightarrow \begin{bmatrix}cos(x_i) & -sin(x_i) \\ sin(x_i) & cos(x_i)\\ \end{bmatrix} := U(x_i)$$
Then our feature vector is encode like so : 
$$S_x = \bigotimes_{i = 1}^NU(x_i)$$
And, remembering the unitary operator $RY(\theta) = \begin{pmatrix} cos\frac{\theta}{2} & -sin \frac{\theta}{2} \\ sin\frac{\theta}{2} & cos \frac{\theta}{2} \\ \end{pmatrix}$, we can use the formula $U(x_i) = RY(2x_i)$ to encode easly every feature.

### Example 
Suppose we want to encode the feature vector $x = (0,-1, 34.56, 0.0001)$ :

```python
x = [0,-1,34.56, 0.0001]
qc = QuantumCircuit(len(x))
for i,xi in enumerate(x):
    qc.ry(xi,i)
qc.draw(output = 'mpl')
```

# Feature maps in Qiskit

## PauliFeatureMap

Qiskit allows implementation of data encoding algorithms that are modeled on different patterns. The first we want to study is the *PauliFeatureMap* : 
$$U_{\Phi(x)} = \prod_dU_{\Phi(x)}H^{\otimes n}$$
with 
$$U_{\Phi(x)} = exp( i\sum_{S\subseteq[n]}\phi_S(x)\prod_{i\in S}P_i )$$
and 
$$\phi_S : x \rightarrow \begin{split} &x_i &\text{ if S = {i}} \\ &(\pi-x_i)(\pi-x_j) &\text{ if S = {i,j}} \\ &\prod_{j \in S}(\pi-x_j) &\text{ otherwise} \end{split}$$
$S \subseteq [n]$ is to be interpreted as $S = [0]$ then $S = [0,1]$, ... . Thus, $i\in S$ means $i = 0$, then $i = (0,1)$,... , leading to a better understanding of the definition of $\phi_S$.

$P_i$ can be any Pauli gate $\{I,X,Y,Z\}$.

The parameter $d$ of the first product is the number of time this circuit is repeated.

Qiskit gives us an implementation in its circuit library using the *ZFeatureMap* and *ZZFeatureMap* classes. These circuits encode $n$ features on $n$ qubits.

```python
from qiskit.circuit.library import ZFeatureMap, ZZFeatureMap
display("ZFeatureMap circuit : ")#Case when n = 0, then S = [0] and P_0 = Z 
                                 #(can be modified to use another gate)
#The first argument is the amount of features we want to encode
qcZ = ZFeatureMap(3,reps = 1, insert_barriers=True)
display(qcZ.decompose().draw(output = 'mpl'))
display("ZZFeatureMap circuit : ")#Case when n = 1, then S = [0,1], P_0 = Z and P_0,1 = ZZ
qcZZ = ZZFeatureMap(3,reps = 1,insert_barriers=True)
display(qcZZ.decompose().draw(output = 'mpl'))
```

We can then send parameters to the circuit to encode our feature vector.  
Suppose that we want to encode the feature vector $x = (1,-2,0.3)$ : 

```python
x = [1,-2,0.3]
encode = qcZZ.bind_parameters(x)#The bind parameters method allows us to insert our values 
                                #in order
encode.decompose().draw(output = 'mpl')
```

## N-local

Another implemented way of encoding data is using the *Nlocal* and *TwoLocal* classes. 
The idea behind these implementation is to encode the data into the rotation parameter of rotation gates around the Z and Y axis. Halfway in the rotations, we entangle every qubits together using any kind of two-qubit gates.

```python
from qiskit.circuit.library import TwoLocal

qc_two = TwoLocal(num_qubits=3, reps=1, rotation_blocks=['ry','rz'],
                entanglement_blocks='cz',
                insert_barriers=True)
qc_two.decompose().draw(output = 'mpl')
```

To use the *NLocal* class, we need to define the rotation and entanglement gates. For example, we define our rotation gates to be alternatively $RY$ and $RZ$ gates and our entanglement gates to be $crx$ gates, applied on 3 consecutive qubits.

```python
from qiskit.circuit.library import NLocal
# rotation block:
rot = QuantumCircuit(2)
params = ParameterVector('r', 2)
rot.ry(params[0], 0)#Alternatively ry and rz gates
rot.rz(params[1], 1)

# entanglement block:
ent = QuantumCircuit(4)
params = ParameterVector('e', 3)
ent.crx(params[0], 0, 1)#crx on 3 consecutive gates
ent.crx(params[1], 1, 2)
ent.crx(params[2], 2, 3)

qc_nlocal = NLocal(num_qubits= 4, rotation_blocks=rot,
                   entanglement_blocks=ent, entanglement= 'linear',
                    insert_barriers=True)
qc_nlocal.decompose().draw(output = 'mpl')
```

# Concrete use

## MNIST dataset reduced
To observe an example of data encoding using real data, we will use the reduced MNIST dataset (https://www.kaggle.com/datasets/mohamedgamal07/reduced-mnist?resource=download) that contains images of handwritten digits, in a 28 by 28 pixels format. To use any feature map that we presented before, we must reduce the amount of feature we want to encode. Indeed, 28 by 28 pixels is 784 features to encode, which exceeds the computing capacity of the machine this presentation has been written on. Thus, we will extract from the original image a 7 by 7 image by taking the mean of 4 pixels.

```python
import pathlib
import imageio
import numpy as np
import matplotlib.pyplot as plt

#We retrieve the path of the images
training_paths = pathlib.Path('MNIST/Reduced MNIST Data/Reduced Training data').glob('0/*.jpg')

im_path = list(training_paths)
#We select the second image to be encoded
read = imageio.imread(str(im_path[2])) 


def reduce(array4):
    #A function to make return an array containing the mean of a
    #2*2 array
    return np.array([np.mean(array4)])

reduced = np.zeros((int(read.shape[0]/4),int(read.shape[1]/4)))

for i in range(0,read.shape[0], 4):
    for j in range(0,read.shape[1],4):
        reduced[int(i/4),int(j/4)] = reduce(read[i:i+4,j:j+4])
#We put the reduced data into a list for it to be easly given
#to the bind_parameters method
reduced_img = reduced
reduced = reduced.reshape(49,).tolist()
```

```python
f, axarr = plt.subplots(1,2)
axarr[0].imshow(read,interpolation = 'nearest', cmap = 'gray', )
axarr[0].set_title("Before reduction")
axarr[1].imshow(reduced_img, interpolation = 'nearest', cmap = 'gray')
axarr[1].set_title("After reduction")
print()
```

## Using a ZZFeatureMap
We are using the ZZFeatureMap to encode the real number that represents each pixel (in grayscale). Note that we could use the ZFeatureMap as well, with the same result in term of qubit amount.

```python
#We create a ZZFeatureMap that will be able to host 49 parameters
qc_reduced_ZZ = ZZFeatureMap(49,reps = 1, insert_barriers=True)
qc_reduced_ZZ.bind_parameters(reduced)
#Uncommenting the following line makes the computation of thix ell way slower (3 min)
# qc_reduced_ZZ.decompose().draw(output = 'mpl', scale = 0.1 ) 
```

```python
#We display the amount of gates used and the number of qubits
print("Number of qubits : ",qc_reduced_ZZ.num_qubits)
print("Gates repartition : ",qc_reduced_ZZ.decompose().count_ops())
```

## Using the EfficientSU2 feature map
The EfficientSU2 feature map was created with the intent of minimizing the amount of qubits used and the depth of the circuit. The EfficientSU2 circuit is built on the NLocal model we spoke about earlier.  

```python
from qiskit.circuit.library import EfficientSU2

circ = EfficientSU2(num_qubits= 5, reps=4)
circ.decompose().draw(output='mpl')
```

## An inconvenience
One of the inconvenience of the EfficientSU2 feature map is that you can't control the amount of feature you can encode. Indeed, the constructor of the EfficientSU2 objects takes a qubit amount and a number of repetitions as parameters. 


The formula to find the amount of feature an EfficientSU2 circuit can encode is : 
$$ p = q(2+2r)$$
with $p$ the amount of parameter the circuit can encode, $q$ the amount of qubits used and $r$ the amount of repetitions of the circuit.

In our case, the amount of repetitions would be : $r = \frac{49-2q}{2q}$, however $49-2q$ is necessarly odd and $2q$ is even, thus $r$ cant be an integer and it doenst exist a solution in which we can encode exactly 49 features.


As having a leftover parameter with no feature to give it can have an impact on our training and thus on the fidelity of our model, we can give this parameter an already existing feature. For example, if we want to encode an image of $7*7$ pixels, and our EfficientSU2 feature map allows 50 parameters, the leftover parameter could be the grayscale value of the first pixel of the image, also encoded in the first parameter of the feature map. 

```python
reduced_fitted = reduced.copy()
#We add the first feature to the last parameter
reduced_fitted.append(reduced[0])
binded = circ.bind_parameters(reduced_fitted)
display(binded.decompose().draw(output = 'mpl'))
print("Number of qubits : ",binded.num_qubits)
print("Gates repartition : ",binded.decompose().count_ops())
```
