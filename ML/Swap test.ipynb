# Swap test 
The swap test is a mean to retieve the inner product of two encoded vectors, that we will call $|\psi_a\rangle$ and $|\psi_b\rangle$. These vectors must be encoded with amplitude encoding. 
Thanks to the swap test, we will be able to retrieve the value $|\langle\psi_a|\psi_b\rangle|^2$.  
Hence we start with a circuit in state $|0\rangle|\psi_a\rangle|\psi_b\rangle$, with $|0\rangle$ being an ancilla qubit we will measure the inner product's value from. For the sake of the example, we will encode two vectors : $\psi_a = (1,2,3,4)$ and $\psi_b = (5,6,7,8)$ using amplitude encoding.

```python
psi_a = [1,2,3,4]
psi_b = [5,6,7,8]

fmpsi_a = RawFeatureVector(4)
fmpsi_a = fmpsi_a.bind_parameters(psi_a)
fmpsi_b = RawFeatureVector(4)
fmpsi_b = fmpsi_b.bind_parameters(psi_b)
instpsi_a = fmpsi_a.to_instruction()
instpsi_b = fmpsi_b.to_instruction()

qc = QuantumCircuit(instpsi_a.num_qubits+instpsi_b.num_qubits+1,1)
qc.append(instpsi_a, range(1,instpsi_a.num_qubits+1))
qc.append(instpsi_b, range(instpsi_a.num_qubits+1, instpsi_a.num_qubits+instpsi_b.num_qubits+1))
qc.draw(output = 'mpl')
```

Then we apply an hadamard gate on the ancilla, thus the state is now $\frac{1}{\sqrt{2}}(|0\rangle+|1\rangle)|\psi_a\rangle|\psi_b\rangle$

```python
qc.h(0)
qc.draw(output = 'mpl')
```

We can now use a conditionned swap gate to transform the last state to $\frac{1}{\sqrt{2}}(|0\psi_a\psi_b\rangle + |1\psi_b\psi_a\rangle)$

```python
qc.cswap(0,1,3)
qc.cswap(0,2,4)
qc.draw(output = 'mpl')
```

Applying another hadamard gate to the ancilla qubits puts our circuit in state  
$$\frac{1}{2}|0\rangle\otimes(|\psi_a\psi_b\rangle+|\psi_b\psi_a\rangle) + \frac{1}{2}|1\rangle\otimes(|\psi_a\psi_b\rangle-|\psi_b\psi_a\rangle)$$

```python
qc.h(0)
qc.draw(output = 'mpl')
```

Now we can measure the ancilla qubit. If we measure the probability to have our ancilla in state 0 (that we will call $p_0$), we then have the following formula : 
$$p_0 = \frac{1}{2}+\frac{1}{2}|\langle\psi_a|\psi_b\rangle|^2$$
And we can deduce the absolute value of our inner product : 
$$|\langle\psi_a|\psi_b\rangle| = \sqrt{2p_0-1}$$

```python
qc.measure(0,0)
qc.draw(output = 'mpl')
```

```python
backend = AerSimulator(method = 'density_matrix')
shots = 8192

transpiled = transpile(qc,backend)
result = backend.run(transpiled, shots = shots).result()
plot_histogram(result.get_counts())
```

```python
#Select the probabilities for the states that end with 0 (ancilla at 0) and sum them together.
#With only one measurement, this computation is equivalent to p_0 = result.get_counts()['0']/shots
p_0 = sum(np.array([v for k,v in zip(result.get_counts().keys(),result.get_counts().values()) if k[-1] == '0']))/shots
print("sqrt = ",math.sqrt(2*p_0-1))
```
